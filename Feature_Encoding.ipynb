Benefits of Target Encoding:-
It is a simple and quick encoding technique that doesnt add to the dimensionality of dataset. It may be used as good first try technique.

Limitations of TE:-
It is dependent on target distribution which means target encoding requires careful validation as it can be prone to overfitting.
