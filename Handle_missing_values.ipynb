```python
df=pd.read_csv('/content/drive/MyDrive/Churn_Modelling.csv')
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import MinMaxScaler
```

```python
df.describe().round(2)
```

```python
df.head()
```

```python
new_df=pd.DataFrame(df,columns=['Age','Tenure'])
```

```python
new_df.head()
```

```python
new_df['Age']=new_df['Age'].fillna(new_df['Age'].mean())
```

```python
new_df.info()
```

```python
scaler=MinMaxScaler() #Instantiating
normalised_df=scaler.fit_transform(new_df)
print(normalised_df)
```

```python
# import numpy as np
x_array=np.array([[2],[3],[5],[6],[6]])
scaler=MinMaxScaler()
normalised_arr=scaler.fit_transform(x_array)
print(normalised_arr)
```

```python
x_array=np.array([[2],[3],[5],[6],[6]])
scaler=StandardScaler()
standard_arr=scaler.fit_transform(x_array)
print(standard_arr)
```
